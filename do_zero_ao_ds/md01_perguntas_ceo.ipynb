### 4. Qual são as 10 casas mais caras e a média de seus atributos ?

```python
#ordenar as casas pelo maior preço, selecionar as 10 primeiras, resetar o index, dropar o antigo e selecionar os atributos desejados
top_ten_houses = data.sort_values(by='price', ascending=False).reset_index(drop=True).head(10)
```

```python
top_ten_houses.head(1)
```

```python
# Calculando a média das colunas selecionadas e colocando em um novo DataFrame
top_ten_mean = pd.DataFrame(top_ten_houses[['bedrooms', 'price', 'bathrooms', 'sqft_living', 'floors', 'grade', 'condition']].mean()).T
```

```python
# Mostrando o DataFrame com as médias de cada atributo entre as 10 casas mais caras
top_ten_mean
```

```python
for i, id in enumerate(top_ten_houses['id']):
    print(f"A {i+1}ª Casa mais cara é a de id {id}, que custa R${top_ten_houses.loc[top_ten_houses['id'] == id, 'price'].to_string(index=False)}, possui {top_ten_houses.loc[top_ten_houses['id'] == id, 'bathrooms'].to_string(index=False)} banheiros e {top_ten_houses.loc[top_ten_houses['id'] == id, 'bedrooms'].to_string(index=False)} quarto(s)") # Adicionar depois "e sua condição é: " quando criar a categorização da condição com base nos números
```

```python
 for i, id in enumerate(bottom_ten_houses['id']):
    print(f"A {i+1}ª Casa mais barata é a de id {id}, que custa R${bottom_ten_houses.loc[bottom_ten_houses['id'] == id, 'price'].to_string(index=False)} possui {bottom_ten_houses.loc[bottom_ten_houses['id'] == id, 'bathrooms'].to_string(index=False)} banheiros e {bottom_ten_houses.loc[bottom_ten_houses['id'] == id, 'bedrooms'].to_string(index=False)} quarto(s)") # Adicionar depois "e sua condição é: " quando criar a categorização da condição com base nos números
```

### Crie um gráfico de barra comparando a média dos atributos das 10 casas mais caras e das 10 casas mais baratas

```python
# Concatenando as duas colunas 
top_bottom_mean = pd.concat([top_ten_mean, bottom_ten_mean], ignore_index=True)
top_bottom_mean
```

```python
# Criando uma coluna que especifique se a linha corresponde às 10 casas mais caras ou às 10 casas mais baratas. (Só consegui plotar no seaborn dessa...
#... forma pois ele não lê o index 0 e 1)

#Criando a nova coluna
top_bottom_mean['price_category'] = 0

# A primeira linha (0) corresponde às médias dos atributos das 10 casas mais caras, portanto recebe o valor 'top_expensive'
top_bottom_mean['price_category'].iloc[0] = 'top_expensive'

# A segunda linha (1) correspondes às médias dos atributos dos 10 casas mais baratas, portanto recebe o valor 'top_cheapest' 
top_bottom_mean['price_category'].iloc[1] = 'top_cheapest'
```

```python
top_bottom_mean.columns
```

```python
#Criando uma grade de gráficos com duas linhas e quatro colunas
fig, axes = plt.subplots(2, 4, figsize=(18, 10))

# Escolhendo o título da grade
fig.suptitle('Comparação entre das médias dos atributos entre as 10 casas mais caras e as 10 casas mais baratas ')



# Plotando os gráficos em suas posições
sns.barplot(ax=axes[0, 0], x='price_category', y='bedrooms', data=top_bottom_mean).set(xlabel=None)
sns.barplot(ax=axes[0, 1], x='price_category', y='bathrooms', data=top_bottom_mean ).set(xlabel=None)
sns.barplot(ax=axes[0, 2], x='price_category', y='price', data=top_bottom_mean).set(xlabel=None)
sns.barplot(ax=axes[0, 3], x='price_category', y='sqft_living', data=top_bottom_mean).set(xlabel=None)
sns.barplot(ax=axes[1, 0], x='price_category', y='floors', data=top_bottom_mean).set(xlabel=None)
sns.barplot(ax=axes[1, 1], x='price_category', y='grade', data=top_bottom_mean).set(xlabel=None)
sns.barplot(ax=axes[1, 2], x='price_category', y='condition', data=top_bottom_mean).set(xlabel=None);
```
